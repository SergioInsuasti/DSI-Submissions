```python
import pandas as pd

%matplotlib inline
```

<div class="alert alert-block alert-info">
<b>Read Seek cleaned data into a DataFrame, ONLY select Category and Description</b> 
</div>

```python
df = pd.read_csv('Datasets/SEEK_Cleaned.csv', usecols=['Unnamed: 0','category', 'description'],index_col=0)

df.head()
```

<div class="alert alert-block alert-info">
<b>Plot Job Category and counts</b> 
</div>

```python
df.category.value_counts().plot.bar(figsize=(16,9))
```

<div class="alert alert-block alert-info">
<b>Use Category and split by Category counts</b> 
</div>

```python
df.dropna(inplace=True)                              # Remove NaN values
df.isnull().sum()                                    # Make sure there are no nulls
```

```python
keep =   ['Data-Scientist',
          'Data-Science',
          'Data-Analyst',
          'Data-Engineer',
          'Data-Architect',
          'Research-Scientist',
          'Business-Intelligence Data',
          'BI-Data',
          'Business-Analyst-Data',
          'BA-Data']


df = df[df['category'].isin(keep)]
df.shape

df.category.replace('"BA-Data"','"Business-Analyst-Data"', inplace=True)
df.category.replace('"BI-Data"','"Business-Intelligence-Data"', inplace=True)
df.category.replace('"Data-Science"','"Data-Scientist"', inplace=True)

df.category.value_counts()
```

```python
df.head()
```

<div class="alert alert-block alert-info">
<b>Use Tfid and Vectorize Categories</b> 
</div>

```python
import sklearn.linear_model
import sklearn.feature_extraction

tfidf = sklearn.feature_extraction.text.TfidfVectorizer()

vectors = tfidf.fit_transform(df.description)

vectors.shape
```

<div class="alert alert-block alert-info">
<b>Create New DataFrame with Tfidf sorted MEAN data</b> 
</div>

```python
New_df = pd.DataFrame(vectors.todense(),columns=tfidf.get_feature_names(), index=df.category)

sorted(zip(New_df[New_df.index == 'Data Scientist'].mean(),df.columns), reverse=True)
New_df.head(50)
```

<div class="alert alert-block alert-info">
<b>Covert Vectors to array</b> 
</div>

```python
vectors.toarray()
```

```python
New_df[New_df.index == 'Data Scientist']
```

<div class="alert alert-block alert-info">
<b>Use Logistic Regression and Fit the Model</b> 
</div>

```python
lr = sklearn.linear_model.LogisticRegression()

lr.fit(vectors, df.category)
```

<div class="alert alert-block alert-info">
<b>Check results on all categories</b> 
</div>

```python
sorted(zip(lr.coef_[0], New_df.columns))
```

<div class="alert alert-block alert-info">
<b>Reverse the order</b> 
</div>

```python
sorted(zip(lr.coef_[0], New_df.columns), reverse=True)
```

<div class="alert alert-block alert-info">
<b>Score the Classification</b> 
</div>

```python
import sklearn.model_selection

score = sklearn.model_selection.cross_val_score(lr, vectors, df.category)

print (score.mean())
```

<div class="alert alert-block alert-info">
<b>Train using SGDClassifier</b> 
</div>

```python
(Xtrain, Xtest, Ytrain, Ytest) = sklearn.model_selection.train_test_split(df.description, df.category)

import numpy as np
from sklearn.pipeline import Pipeline
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.feature_extraction.text import TfidfTransformer
from sklearn.linear_model import SGDClassifier

text_clf = Pipeline([('vect', CountVectorizer()),
                     ('tfidf', TfidfTransformer()),
                     ('clf', SGDClassifier(loss='hinge', penalty='l2',
                                           alpha=1e-3, random_state=42,
                                           #max_iter=5, tol=None
                                          )),
                   ])

text_clf.fit(Xtrain, Ytrain)  


predicted = text_clf.predict(Xtest)
np.mean(predicted == Ytest)
```
