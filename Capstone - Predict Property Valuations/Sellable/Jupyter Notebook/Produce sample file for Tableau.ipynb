<div class="alert alert-block alert-info">
<b>Stplit Address into street_no, street_name, street_suffix</b> 
</div>

```python
def strip_until_space(field):
    new_field1 = []
    new_field2 = []
    flag = 'N'           
    for c in field:
        if (c == ' ' or c == '' or c == '') and flag == 'N':
            flag = 'Y'
        if flag == 'Y':
            new_field1.append(c)
        else:
            new_field2.append(c) 
    split_field1 = ''.join(new_field1[1:])
    split_field2 = ''.join(new_field2[:])
    return split_field1, split_field2
```
