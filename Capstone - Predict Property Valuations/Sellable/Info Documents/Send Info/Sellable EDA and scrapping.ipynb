```python
def find_house_info(i, count, hold_id):
    driver.get('https://www.realestateview.com.au/property-360/property/' + i)
    
    dd = ('https://www.realestateview.com.au/property-360/property/' + i)
    
    print (str(count), '\t', hold_id, '\t', dd)
    #grab the page source
    html = driver.page_source
    
    # BeautifulSoup it!
    html = BeautifulSoup(html, 'lxml')
    
    data_value = []
    
    # Extract information
    extract_house_info(html, data_value, hold_id)
```

```python
def extract_house_info(html, data_value, hold_id):
#    print ('UUUUUUUU-->\t', hold_id)
    sale_method_1 = ' '
    sale_method_2 = ' '
    value_zero = 0
    value_space = ' '
    r_data = []
    data_value.append(hold_id)   
# Rooms Beds Cars Lot
    for data in html.find_all('div', {'class':'rooms'}):
        r_data.append(data.text)
        word = '\n'.join(r_data[0:3])
        r_data = word.split('\n')

        while '' in r_data:
            r_data.remove('')
        count = 0    
        for x in r_data:   
            try:
                data_value.append(re.findall(r'\d+', x)[0])
            except:
                if IndexError:
                    data_value.append(value_zero)
            count += 1
            
        count = 4 - count 
#        print ('Count is\t', count)
        while count > 0:
            data_value.append(value_zero)
            count = count - 1
#            print ('count', count)
       
        hold_data = data_value
 
        break
#    print ("1++++", hold_data)
#    return hold_data
    
# Valuation
    r_data = []
    for data in html.find_all('div', {'class':'top-3comps'}):
        r_data.append(data.text)
        word = '\n'.join(r_data[0:3])
        r_data = word.split('\n')
      
        while '' in r_data:
            r_data.remove('')
        break
    try:
        hold_data.append(r_data[3])
    except:
        if IndexError:
            hold_data.append(value_space)

#    print ("2++++", hold_data)

# Sale Method
    r_data = []
    for data in html.find_all('div', {'col-xs-6 col-sm-3 col-md-3 col-lg-3 sale-method'}):
        r_data.append(data.text)
        word = '\n'.join(r_data[0:5])
        r_data = word.split('\n')
    
    while '' in r_data:
        r_data.remove('')

    try:
        if r_data[1] != '':
            auction_sale_1 = r_data[1].strip()
    except:
        if IndexError:
            auction_sale_1 = value_space
        
    try:
        if r_data[2] != '':
            auction_sale_2 = r_data[2].strip()
    except:
        if IndexError:
            auction_sale_2 = value_space
        
    if auction_sale_1 == 'Auction' or auction_sale_1 == 'Private Sale':
        hold_data.append(r_data[1])
        sale_method_1 = r_data[1]
    else:
        hold_data.append(value_space)
    
    if auction_sale_2 == 'Auction' or auction_sale_2 == 'Private Sale':
        hold_data.append(r_data[2])
        sale_method_2 = r_data[2]
    else:
        hold_data.append(value_space)
        
#    print ("3++++", hold_data)
#    print ('r_data\t', r_data)
    
# Date Sold
    r_data = []
    for data in html.find_all('div', {'class':'col-xs-6 col-sm-2 col-md-2 col-lg-2 date'}):
        r_data.append(data.text)
        word = '\n'.join(r_data[0:5])
        r_data = word.split('\n')
    
    while '' in r_data:
        r_data.remove('')
        
    if auction_sale_1 == 'Auction' or auction_sale_1 == 'Private Sale':
        hold_data.append(r_data[1])
    else:
        hold_data.append(value_space)
        
    if auction_sale_2 == 'Auction' or auction_sale_2 == 'Private Sale':
        hold_data.append(r_data[2])
    else:
        hold_data.append(value_space)
        
#    print ("4++++", hold_data)
#    print ('r_data\n', r_data)
    
# Sold Price
    r_data = []
    for data in html.find_all('div', {'class':'col-xs-4 col-sm-2 col-md-2 col-lg-3 price'}):
        r_data.append(data.text)

        word = '\n'.join(r_data[0:5])
        r_data = word.split('\n')
    
    while '' in r_data:
        r_data.remove('')
    if auction_sale_1 == 'Auction' or auction_sale_1 == 'Private Sale':
        try:
            hold_data.append(r_data[1])
        except:
            if IndexError:
                data_value.append(value_space)
    else:
        hold_data.append(value_space)
        
    if auction_sale_2 == 'Auction' or auction_sale_2 == 'Private Sale':
        try:
            hold_data.append(r_data[2])
        except:
            if IndexError:
                data_value.append(value_space)
    else:
        hold_data.append(value_space)
#    print ("5++++", hold_data)
#    print ('r_data\n', r_data)
    
# House Blog
    r_data = []
    for data in html.find_all('p', {'class':'magigraph-top'}):
        r_data.append(data.text)

        word = '\n'.join(r_data[0:3])
        r_data = word.split('\n')
    
    while '' in r_data:
        r_data.remove('')

    hold_data.append(r_data[0])
#    print ("6++++", hold_data)
#    print (hold_data)
    datalist.append(hold_data)
    return hold_data
```
